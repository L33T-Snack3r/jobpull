### Accessing postgres db

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",        # or 127.0.0.1
    port=5432,
    dbname="jobs_db",
    user="postgres",
    password="postgres"
)

cur = conn.cursor()
```

```python
jobs.columns
```

```python
jobs_df = jobs[['id', 'date_posted', 'job_url', 'job_url_direct', 'title',
                'company', 'company_description', 'description', 'company_industry',
                'company_url', 'company_num_employees', 'site'
                ]]
```

```python
q = """                              
SELECT column_name, data_type, is_nullable
FROM information_schema.columns
WHERE table_name = %s;
"""

cur = conn.cursor()
cur.execute(q, ('joblistings',))  # (table_name,) passed as tuple
cur.fetchall()
```

```python
q = (f"INSERT INTO joblistings "
     "(job_id, date_posted, job_url, job_url_direct, title, "
     "company, company_description, job_description, company_industry, "
     "company_url, company_num_employees, website)"
     " VALUES "
     "(%s, %s, %s, %s, %s,"
     "%s, %s, %s, %s,"
     "%s, %s, %s)")

cur = conn.cursor()
cur.executemany(q, jobs_df.values.tolist())
```

```python
cur.execute("""SELECT table_name FROM information_schema.tables
       WHERE table_schema = 'public'""")
```

```python
cur.execute("SELECT * FROM joblistings LIMIT 5;")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```
